```python
import pandas as pd
import matplotlib.pyplot as plt

datos = pd.read_csv('casasboston.csv')
datos
```

```python
df = datos[['RM', 'CRIM', 'MEDV', 'TOWN','CHAS']]

df=datos.rename(columns={
    'TOWN':'CIUDAD',
    'CRIM':'INDICE_CRIMEN',
    'INDUS':'PCT_ZONA_INDUSTRIAL',
    'CHAS':'RIO CHARLES',
    'RM':'N_HABITACIONES_MEDIO',
    'MEDV':'VALOR_MEDIANO',
    'LSTAT':'PCT_CLASE_BAJA'
})
df
```

```python
df.N_HABITACIONES_MEDIO.plot.hist(column=['Numero de Habitaciones'])
plt.show()
```

```python
df.plot.scatter(x='VALOR_MEDIANO', y='N_HABITACIONES_MEDIO')
plt.show()
```

```python
valor_por_ciudad=df.groupby('CIUDAD')['VALOR_MEDIANO'].mean()
valor_por_ciudad.head(10).plot.barh()
plt.show()
```
