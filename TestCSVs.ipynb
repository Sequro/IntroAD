```python
#Â¿CuÃ¡les son los aÃ±os con el mayor y menor uso de tierras agrÃ­colas en Costa Rica?
agricultural = data[data['Indicator Name'] == "Agricultural land (sq. km)"]

max_agricultural_year = agricultural.loc[agricultural['Value'].idxmax()][['Year', 'Value']]
print("Mayor uso de tierras agricolas:\n",max_agricultural_year[['Year', 'Value']])
min_agricultural_year = agricultural.loc[agricultural['Value'].idxmin()][['Year', 'Value']]
print("Menor uso de tierras agricolas:\n",min_agricultural_year[['Year', 'Value']])
```

```python
# Area forestal promedio en Costa Rica
promedio_forest = data[data['Indicator Name'] == 'Forest area (sq. km)']['Value']
forest_data = pd.to_numeric( promedio_forest, errors='coerce')
print('\n El promedio de area forestal en Costa Rica es de:', forest_data.mean())
```
