```python
import pandas as pd
import matplotlib.pyplot as plt

df=pd.read_csv('/content/climate-change_cri.csv')

data=pd.DataFrame(df)
data
```

```python
#Indicadores dentro del cambio climatico en Costa Rica
indicators=data['Indicator Name'].unique()
indicators
```

```python
# filtrar los datos del area forestal de 1960 y 2023 para comparar si las mimas ha aumentado o disminuido
forest_1960 = data.loc[(data['Indicator Name'] == 'Forest area (sq. km)') & (data['Year'] >= 1960) & (data['Year']<=2000), 'Value']
forest_2023 = data.loc[(data['Indicator Name'] == 'Forest area (sq. km)') & (data['Year'] >= 2001) & (data['Year']<=2023), 'Value']

if forest_1960.mean() > forest_2023.mean():
  print("El area forestal en Costa Rica ha disminuido desde 1960 a 2023")
else:
  print("El area forestal en Costa Rica ha aumentado desde 1960 a 2023")

```
