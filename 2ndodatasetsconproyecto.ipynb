```python
import pandas as pd
import matplotlib.pyplot as plt

df=pd.read_csv('sample_data/climate-change_cri.csv')

data=pd.DataFrame(df)
data
```

```python
import pandas as pd
df['Value'] = pd.to_numeric (df['Value'], errors= 'coerce')
registro_emisiones= df.loc[(df['Indicator Name'].isin(['CO2 emissions from liquid fuel consumption (kt)','CO2 emissions from solid fuel consumption (kt)'])) & (df['Value'].notnull())]
promedio_emisiones = registro_emisiones.groupby(['Indicator Name'])['Value'].mean().to_frame()
promedio_emisiones= promedio_emisiones.round(2)
promedio_emisiones = promedio_emisiones.rename(columns={'Value': 'Promedio de emisiones de CO2'})
print(f'Promedio: \n: ', promedio_emisiones)
```
