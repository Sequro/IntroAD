```python
import pandas as pd
import matplotlib.pyplot as plt

df=pd.read_csv('sample_data/climate-change_cri.csv')

data=pd.DataFrame(df)
data
```

```python
#Indicadores dentro del cambio climatico en Costa Rica
indicators=data['Indicator Name'].unique()
indicators
```

```python
# Datos del año 2019.
datos_2019 = data.loc[(data['Year'] == 2019)]
datos_2019
```

```python
# Indicador "CO2 emissions (kt)" para 2018.
agricultural_2018 = data.loc[(data['Year'] == '2018') & (data['Indicator Name'] == "CO2 emissions (kt)")]
agricultural_2018
```

```python
# Datos de Costa Rica antes de 2017.
costa_rica = data.loc[(data['Country Name'] == "Costa Rica")]
costa_rica_antes_2017 = costa_rica.loc[(costa_rica['Year'] < 2017)]
costa_rica_antes_2017
```

```python
# Datos con el indicador "Agricultural land (sq. km)".
data_agri = data.loc[(data['Indicator Name'] == "Agricultural land (sq. km)")]
data_agri
```

```python
# Tendencia de "Agricultural land (sq. km)" para Costa Rica entre 2015 y 2020.
trend_2015_2020 = data_agri.loc[(data_agri['Year'] >= '2015') & (data_agri['Year'] <= '2020')][['Year', 'Value']]
trend_2015_2020
```

```python
# area forestal promedio en Costa Rica
promedio_forest = data[data['Indicator Name'] == 'Forest area (sq. km)']['Value']
forest_data = pd.to_numeric( promedio_forest, errors='coerce')
forest_data.mean()
```

```python
# año mínimo y máximo en el CSV
df['Year'] = pd.to_numeric(df['Year'], errors='coerce')
year_range = df['Year'].min(), df['Year'].max()
print(f"El rango de años en el CSV es: {year_range[0]} - {year_range[1]}")
```

```python
#Comparacion de cada 10 años sobre todos los indicadores del CO2
print('\n Comparacion del CO2 cada 10 años \n')
print(' \n Año 1990 ')
dataCO2_80=data.loc[((data['Indicator Name'].str.contains('CO2')) & (data['Year']==1990))]
print(dataCO2_80[[ 'Year', 'Indicator Code','Value']])

print(' \n Año 2000')
dataCO2_00=data.loc[((data['Indicator Name'].str.contains('CO2')) & (data['Year']==2000))]
print(dataCO2_00[[ 'Year', 'Indicator Code','Value']])

print(' \n Año 2010')
dataCO2_10=data.loc[((data['Indicator Name'].str.contains('CO2')) & (data['Year']==2010))]
print(dataCO2_10[[ 'Year', 'Indicator Code','Value']])

print(' \n Año 2020')
dataCO2_20=data.loc[((data['Indicator Name'].str.contains('CO2')) & (data['Year']==2020))]
print(dataCO2_20[[ 'Year', 'Indicator Code','Value']])
```

```python
#Comparacion del porcentaje del cosumo de energia renobable cada 10 años
print('\n Comparacion del porcentaje del cosumo de energia renobable cada 10 años \n')
print(' \n Año 1990 ')
datarenewcomp90=data.loc[(data['Indicator Name']=='Renewable energy consumption (% of total final energy consumption)')&(data['Year']==1990)]
print(datarenewcomp90[[ 'Year', 'Indicator Code','Value']])

print(' \n Año 2000 ')
datarenewcomp00=data.loc[(data['Indicator Name']=='Renewable energy consumption (% of total final energy consumption)')&(data['Year']==2000)]
print(datarenewcomp00[[ 'Year', 'Indicator Code','Value']])

print(' \n Año 2010 ')
datarenewcomp10=data.loc[(data['Indicator Name']=='Renewable energy consumption (% of total final energy consumption)')&(data['Year']==2010)]
print(datarenewcomp10[[ 'Year', 'Indicator Code','Value']])

print(' \n Año 2020 ')
datarenewcomp20=data.loc[(data['Indicator Name']=='Renewable energy consumption (% of total final energy consumption)')&(data['Year']==2020)]
print(datarenewcomp20[[ 'Year', 'Indicator Code','Value']])
```

```python
# Toda la informacion del año 1960
data60=data.loc[(data['Year']==1960)]
print('Costa Rica anno 1960 \n')
print(data60[[ 'Year', 'Indicator Code','Value']])

data20=data.loc[(data['Year']==2020)]
print('Costa Rica anno 2020 \n')
print(data20[[ 'Year', 'Indicator Code','Value']])

# Porcentaje de crecimiento de la poblacion del 1980 a 2010
print('\n Porcentaje crecimiento de poblacion en Costa Rica \n')
dataPop=data.loc[((data['Indicator Name']=='Population growth (annual %)') & (data['Year']>1980) & (data['Year']<2010))]
print(dataPop.sort_values(by=['Value'], ascending=True)[['Country Name', 'Year', 'Value']])
```
