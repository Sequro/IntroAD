```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('/content/climate-change_cri.csv')
df = pd.DataFrame(data)
```

```python
relevant_indicators = [
    'Renewable electricity output (% of total electricity output)',
    'Access to electricity (% of population)',
    'Electricity production from coal sources (% of total)',
    'Electricity production from hydroelectric sources (% of total)',
]

filtered_df = df[df['Indicator Name'].isin(relevant_indicators)]
filtered_df['Value'] = pd.to_numeric(filtered_df['Value'], errors='coerce')
```

```python
contingency_table = pd.crosstab(filtered_df['Year'], filtered_df['Indicator Name'], values=filtered_df['Value'], aggfunc='mean')
plt.figure(figsize=(10, 6))
sns.heatmap(contingency_table, annot=True, cmap="Set3")
plt.title('Mapa de Calor: Indicadores Clave vs Ao')
plt.show()
```
