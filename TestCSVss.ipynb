```python
import pandas as pd
import matplotlib.pyplot as plt

df=pd.read_csv('/content/climate-change_cri.csv')

data=pd.DataFrame(df)
data
```

```python
#Indicadores dentro del cambio climatico en Costa Rica
indicators=data['Indicator Name'].unique()
indicators
```

```python
# Datos del año 2019.
datos_2019 = data.loc[(data['Year'] == 2019)]
datos_2019
```

```python
# Indicador "CO2 emissions (kt)" para 2018.
agricultural_2018 = data.loc[(data['Year'] == '2018') & (data['Indicator Name'] == "CO2 emissions (kt)")]
agricultural_2018
```

```python
# Datos de Costa Rica antes de 2017.
costa_rica = data.loc[(data['Country Name'] == "Costa Rica")]
costa_rica_antes_2017 = costa_rica.loc[(costa_rica['Year'] < '2017')]
costa_rica_antes_2017
```

```python
# Datos con el indicador "Agricultural land (sq. km)".
data_agri = data.loc[(data['Indicator Name'] == "Agricultural land (sq. km)")]
data_agri
```

```python
# Tendencia de "Agricultural land (sq. km)" para Costa Rica entre 2015 y 2020.
trend_2015_2020 = data_agri.loc[(data_agri['Year'] >= '2015') & (data_agri['Year'] <= '2020')][['Year', 'Value']]
trend_2015_2020
```

```python
# area forestal promedio en Costa Rica
promedio_forest = data[data['Indicator Name'] == 'Forest area (sq. km)']['Value']
forest_data = pd.to_numeric( promedio_forest, errors='coerce')
forest_data.mean()
```

```python
# año mínimo y máximo en el CSV
df['Year'] = pd.to_numeric(df['Year'], errors='coerce')
year_range = df['Year'].min(), df['Year'].max()
print(f"El rango de años en el CSV es: {year_range[0]} - {year_range[1]}")
```

```python
#Niveles de consumo de energia verde

Consume_green_energy=data.loc[(data['Indicator Name'] == 'Renewable energy consumption (% of total final energy consumption)') & (data['Value'])]
print(Consume_green_energy.sort_values(by='Year',ascending=False))

```
